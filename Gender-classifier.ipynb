```python
 df.dropna(subset=['_unit_id', '_trusted_judgments', 'gender:confidence',
       'profile_yn:confidence', 'fav_number', 'retweet_count', 'tweet_count',
       'tweet_id'], inplace=True)
```

```python
X=df[['_unit_id','_trusted_judgments',
       'profile_yn:confidence','tweet_count',
       'tweet_id']].values
Y=df['gender'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y)
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train2=sc.fit_transform(x_train)
```

```python
x_test2=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_train2
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(activation='relu', alpha=0.0001, batch_size='auto', beta_1=0.9,beta_2=0.999, early_stopping=False, epsilon=1e-08,hidden_layer_sizes=(10,), learning_rate='constant',learning_rate_init=0.001, max_fun=15000, max_iter=200,momentum=0.9, n_iter_no_change=100, nesterovs_momentum=True,power_t=0.5, random_state=None, shuffle=True, solver='adam',tol=0.0001, validation_fraction=0.1, verbose=1, warm_start=True)
```

```python
mlp
```

```python
mlp.fit(x_train2,y_train)
```

```python
y_pred=mlp.predict(x_test2)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
